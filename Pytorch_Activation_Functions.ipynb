### Activation Function
---
* Activation functions is used for a Artificial Neural Network to learn and make sense of something really complicated and Non-linear complex functional mappings between the inputs and response variable. They introduce non-linear properties to our Network.
* Their main purpose is to convert a *input signal of a node in a A-NN to an output signal. That output signal now is used as a input in the next layer in the stack.
![](https://i0.wp.com/sefiks.com/wp-content/uploads/2020/02/sample-activation-functions-square.png)

```
import numpy as np
import torch
import torch.nn.functional as F
from torch.autograd import Variable
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings(action='ignore')
```

* Generate the data

```
X = torch.linspace(-5,5, 200) # X data(tensor), shape = (200,1)
X = Variable(X)
X_np = X.data.numpy() # convert data into numpy array
```

Popular Activation Function
---

* **1. Sigmoid**


image.png

```
y_sigmoid = F.sigmoid(X).data.numpy()
# print(f"Sigmoid Output : \n{y_sigmoid}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_sigmoid, c='red', label='Sigmoid')
plt.title('Sigmoid  Activation')
plt.ylim((0, 1))
plt.legend(loc='best')
plt.show()
```

* **2. Tanh**

image.png

```
y_tanh = F.tanh(X).data.numpy()
# print(f"Tanh Output : \n{y_tanh}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_tanh, c='red', label='Sigmoid')
plt.title('Tanh  Activation')
plt.ylim((-1, 1))
plt.legend(loc='best')
plt.show()
```

* **3. Step Function**


image.png

```
def step_function(X):
  if X < 0.5:
    return 0
  elif X >= 0.5:
    return 1
  else:
    raise("your value is not in range")

y_step = []
for i in X:
  y_step.append([step_function(i.numpy())])

y_step = torch.tensor(y_step).view(-1)
# y_step = torch.from_numpy(np.array(y_step))
# print(f"Step Function : \n{y_step.numpy()}")   

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_step, c='red', label='Tanh')
plt.title('Step Function Activation')
plt.ylim((0, 1))
plt.legend(loc='upper left')
plt.show()
```

* **4. SoftPlus**

image.png

```
y_softplus = F.softplus(X).data.numpy()
# print(f"Softplus Output : \n{y_softplus}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_softplus, c='red', label='Softplus')
plt.title('Softplus  Activation')
plt.ylim((0, 1))
plt.legend(loc='best')
plt.show()
```

* **5.Relu**

image.png

```
y_relu = F.relu(X).data.numpy()
# print(f"Relu Output : \n{y_relu}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_relu, c='red', label='Relu')
plt.title('Relu  Activation')
plt.ylim((0, 6))
plt.legend(loc='best')
plt.show()
```

* **6.Soft Sign**

image.png

```
y_softsign = F.softsign(X).data.numpy()
# print(f"SoftSign Output : \n{y_softsign}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_softsign, c='red', label='Softsign')
plt.title('Softsign  Activation')
plt.ylim((-1, 1))
plt.legend(loc='best')
plt.show()
```

* **7.Elu**

image.png

```
y_elu = F.elu(X).data.numpy()
# print(f"Elu Output : \n{y_elu}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_elu, c='red', label='Elu')
plt.title('Elu  Activation')
plt.ylim((-2, 6))
plt.legend(loc='best')
plt.show()
```

* **8.Log of sigmoid**

image.png

```
y_logsigmoid = F.logsigmoid(X).data.numpy()
# print(f"Log Sigmoid Output : \n{y_logsigmoid}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_logsigmoid, c='red', label='Log Sigmoid')
plt.title('Log Sigmoid  Activation')
plt.ylim((-6, 2))
plt.legend(loc='best')
plt.show()
```

* **9.Swish**


image.png




```
y_swish = X * F.sigmoid(X).data.numpy()
# print(f"Swish Output : \n{y_swish}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_swish, c='red', label='Swish')
plt.title('Swish Activation')
plt.ylim((-2, 6))
plt.legend(loc='best')
plt.show()
```

* **10.Sinc**


image.png

```
y_sinc = (torch.sin(X).data.numpy()) / X
# print(f"Sinc Output : \n{y_sinc}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_swish, c='red', label='Sinc')
plt.title('Sinc Activation')
plt.ylim((-1, 1.5))
plt.legend(loc='best')
plt.show()
```

* **11.Leaky Relu**


image.png

```
y_leakyrelu = F.leaky_relu(X).data.numpy()
# print(f"Leaky Relu Output: \n{y_leakyrelu}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_leakyrelu, c='red', label='Leaky relu')
plt.title('Leaky Relu')
plt.ylim((-1, 6))
plt.legend(loc='best')
plt.show()
```

* **12.Mish**


image.png

```
y_mish = X * F.tanh(F.softplus(X)).data.numpy()
# print(f"Mish Output: \n{y_mish}")

plt.figure(figsize=(8, 6))
plt.scatter(X_np, y_mish, c='red', label='Mish')
plt.title('Mish Activation')
plt.ylim((-1, 6))
plt.legend(loc='best')
plt.show()
```
